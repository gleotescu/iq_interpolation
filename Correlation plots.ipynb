# Covariance represention


# Positive correlation

```python
import numpy as np

np.random.seed(1)

# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Positive Correlation with some noise
y = x + np.random.normal(0, 10, 1000)

np.corrcoef(x, y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
plt.clf()
plt.scatter(x, y)
plt.rcParams["figure.figsize"] = [12,8]
plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=12
plt.savefig('positiveCorrelated.png', dpi = 200) 
plt.show()
```

# Negative correlation

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Negative Correlation with some noise
y = 100 - x + np.random.normal(0, 5, 1000)
# plt.title("Negative correlated variable")
np.corrcoef(x, y)
plt.scatter(x, y)
plt.rcParams["figure.figsize"] = [12,8]
plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=12
plt.savefig('negativeCorrelated.png', dpi = 200) 
plt.show()
```

# Near-zero correlated variable

```python
x = np.random.randint(10, 50, 100)
y = np.random.randint(30, 80, 100)
np.corrcoef(x, y)
plt.scatter(x, y)
# plt.title("Near zero correlated variable")
plt.xlim([0,60])
plt.ylim([0,100])
plt.savefig('nearZeroCorrelated.png', dpi = 200)
plt.show()
```
